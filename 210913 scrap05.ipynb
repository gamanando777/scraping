```python
# 관련 모듈 임포트
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

### 한 페이지의 댓글 한개

```python
# 1) 페이지 주소 요청 => 텍스트
url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=167697&target=after&page=1'
response = requests.get(url)
html_str = response.text
html_str
```

```python
# 2) BeautifulSoup 객체화
soup = BeautifulSoup(html_str, 'html.parser') # parser 보기좋게(유의미하게?) 정리
soup
```

```python
# 3) 필터링(반복되는 요소의 선택자)
td_list = soup.find_all(class_ = 'title')
len(td_list)
```

```python
td_list[0]
```

```python
# 첫번째 댓글의 제목, 평점, 내용, url
print(td_list[0].find('a').text) # 제목
print(td_list[0].find('em').text) # 평점
print(td_list[0].find('br').next_sibling.strip()) # 내용
# https://movie.naver.com/movie/bi/mi/basic.naver?code=167697
base_url = 'https://movie.naver.com'
print(base_url+td_list[0].find('a')['href']) # url
```

```python
# 마지막 댓글의 제목, 평점, 내용
print(td_list[-1].find('a').text) # 제목
print(td_list[-1].find('em').text) # 평점
print(td_list[-1].find('br').next_sibling.strip()) # 내용
base_url = 'https://movie.naver.com'
print(base_url+td_list[-1].find('a')['href']) # url
```

```python
# 4) 2차원 리스트로 저장 => 데이터프레임화
movie_review1 = []
base_url = 'https://movie.naver.com'
for td in td_list:
    title = td.find('a').text
    score = td.find('em').text
    review = td.find('br').next_sibling.strip()
    url = base_url + td.find('a')['href']
    movie_review1.append([title, score, review, url])
len(movie_review1)
```

```python
movie_review1[0]
```

```python
df1 = pd.DataFrame(movie_review1, columns = ['title', 'score', 'review','url'])
df1.head(3)
```

## 여러 페이지의 댓글( 페이지 범위 지정)

```python
last_page = 5
movie_review2 = []
base_url = 'https://movie.naver.com'
for i in range(1,last_page+1):
    # 1) 페이지 주소 요청 => 텍스트
    url = f'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=167697&target=after&page={i}'
    response = requests.get(url)
    html_str = response.text
    #html_str

    # 2) BeautifulSoup 객체화
    soup = BeautifulSoup(html_str, 'html.parser') # parser 보기좋게(유의미하게?) 정리
    #soup

    # 3) 필터링(반복되는 요소의 선택자)
    td_list = soup.find_all(class_ = 'title')
    #len(td_list)
    
    # 4) 2차원 리스트로 저장 => 데이터프레임화

    for td in td_list:
        title = td.find('a').text
        score = td.find('em').text
        review = td.find('br').next_sibling.strip()
        url = base_url + td.find('a')['href']
        movie_review2.append([title, score, review, url])
        
  
print(f'전체 개수는? {len(movie_review2)}')
```

```python
df2 = pd.DataFrame(movie_review2, columns = ['title', 'score', 'review', 'url'] )
df2.head(3)
```

```python
df2.tail(3)
```

### 모든 페이지의 댓글

##### 알고리즘
- while True 무한반복문 이용
- 페이지에서 [다음] 버튼 .pg_next 가 없다면 반복문 종료

```python
movie_review3 = []
base_url = 'https://movie.naver.com'
page_num = 0
while True:
    # 1) 페이지 주소 요청 => 텍스트 
    page_num += 1
    url = f'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page={page_num}'
    print(url)
    response = requests.get(url)
    html_str = response.text

    # 2) BeautifulSoup 객체화 
    soup = BeautifulSoup(html_str, 'html.parser')

    # 3) 필터링 (반복되는 요소의 선택자)
    td_list = soup.find_all(class_= 'title')
    
    # 4) 2차원 리스트에 저장 
    for td in td_list:
        title = td.find('a').text
        score = td.find('em').text
        review = td.find('br').next_sibling.strip()
        url = base_url + td.find('a')['href']
        movie_review3.append([title, score, review, url])
        
    # 5) 마지막 페이지면 종료 
    page_next = soup.find_all(class_ = 'pg_next')
    if len(page_next) == 0:
        break

print(f'전체 갯수는? {len(movie_review3)}')
```

```python
df3 = pd.DataFrame(movie_review3, columns = ['title', 'score', 'review', 'url'] )
df3.tail(3)
```

```python
df3.to_csv('output/kid_movie_review.csv', index=False)
```
