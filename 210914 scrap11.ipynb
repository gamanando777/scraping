````python
# clientId = '본인꺼'
# clientSecret = '본인꺼'

clientId = 'jJ01qIEsMvzzAWOZsEuH'
clientSecret = '_7pG1Y7OJm'
````

````python
# 2) API 파이썬 소스 참조 
#  https://developers.naver.com/docs/serviceapi/search/blog/blog.md#%EB%B8%94%EB%A1%9C%EA%B7%B8

# 네이버 검색 API예제는 블로그를 비롯 전문자료까지 호출방법이 동일하므로 blog검색만 대표로 예제를 올렸습니다.
# 네이버 검색 Open API 예제 - 블로그 검색
import os
import sys
import urllib.request
import json
import pandas as pd

# 본인 아이디와 시크릿키로 변경
client_id = clientId
client_secret = clientSecret

# 검색어 지정 후 인코딩
encText = urllib.parse.quote("의정부역 맛집")

# URL 생성
# 요청변수 : display : 10(기본값), 100(최대): 검색 결과 출력 건수 지정
# url에 함께 지정
# http://openapi.naver.com/v1/search/blog?요청변수1=r값1&요청변수2=값2...

# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # json 결과
# display 요청 변수 추가 설정 
url = "https://openapi.naver.com/v1/search/blog?display=100&query=" + encText # json 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과

# 페이지 요청
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)

# 요청 객체 생성 
response = urllib.request.urlopen(request)
rescode = response.getcode()
if(rescode==200):
    #텍스트 데이터
    response_body = response.read()
    #텍스트 디코딩
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
````

````python
blog_data_dict =json.loads(response_body.decode('utf-8'))
type(blog_data_dict)
````

````python
# 실제 필요한 블로그 데이터 리스트 추출
#blog_data_dict
print(type(blog_data_dict['items']))
blog_list = blog_data_dict['items']
print(len(blog_list))
blog_list[0]
````

````python
# 3) 리스트 딕셔너리 => 데이터 프레임
# 빈 데이터프레임 생성
df_blog = pd.DataFrame(columns=['title','link','description','bloggername','bloggerlink','postdate'])
df_blog
````

````python
# 행  => 데이터프레임의 행이름
idx = 1
for dict in blog_list:
    title = dict['title'].replace('<b>','').replace('</b>','')
    link = dict['link']
    description = dict['description'].replace('<b>','').replace('</b>','')
    bloggername = dict['bloggername']
    bloggerlink = dict['bloggerlink']
    postdate = ['postdate']
    # 데이터프레임에 삽입
    df_blog.loc['row'+str(idx)] = [title,link,description,bloggername,bloggerlink,postdate]
    #df_todos.loc[idx] = [userId, id, title, completed]
    idx += 1
````

# 퀴즈 : 네이버 책 검색엔진
- https://developers.naver.com/docs/serviceapi/search/book/book.md#%EC%B1%85

````python
import urllib.request
import json
import pandas as pd
````

````python
    
def book_api():

    clientId = 'jJ01qIEsMvzzAWOZsEuH'
    clientSecret = '_7pG1Y7OJm'

    # 본인 아이디와 시크릿키로 변경
    client_id = clientId
    client_secret = clientSecret

    # 검색어 지정 후 인코딩
    what = input("검색어 입력 => ")
    encText = urllib.parse.quote(f'{what}')

    # URL 생성
    # 요청변수 : display : 10(기본값), 100(최대): 검색 결과 출력 건수 지정
    # url에 함께 지정
    # http://openapi.naver.com/v1/search/kin?요청변수1=r값1&요청변수2=값2...

    # url = "https://openapi.naver.com/v1/search/book?query=" + encText # json 결과
    # display 요청 변수 추가 설정 
    url = "https://openapi.naver.com/v1/search/book?display=100&query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/search/book.xml?query=" + encText # xml 결과

    # 페이지 요청
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)

    # 요청 객체 생성 
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    if(rescode==200):
        #텍스트 데이터
        response_body = response.read()
        #텍스트 디코딩
        response_body.decode('utf-8')
    else:
        print("Error Code:" + rescode)

    book_data_dict =json.loads(response_body.decode('utf-8'))
    book_list = book_data_dict['items']
    #book_list[0]

    # 3) 리스트 딕셔너리 => 데이터 프레임
    # 빈 데이터프레임 생성
    df_book = pd.DataFrame(columns=['title','link','author','publisher','price','discount'])

    # 행  => 데이터프레임의 행이름
    idx = 1
    for dict in book_list:
        title = dict['title'].replace('<b>','').replace('</b>','')
        link = dict['link']
        author = dict['author']
        publisher = dict['publisher']
        price = dict['price']
        discount = dict['discount']
        # 데이터프레임에 삽입
        df_book.loc[idx] = [title,link,author,publisher,price,discount]
        idx += 1

    return df_book

book_api()
````
