# Finance Data Reader

- 국내 주식가격, 주가지수, 환율, 암호화페 데이타 제공 
- https://financedata.github.io/posts/finance-data-reader-users-guide.html
- pip install -U finance-datareader

```python
# pip list
```

```python
# 모듈 임포트 
import pandas as pd
import FinanceDataReader as fdr
```

### 주가 데이타 

- 대표적인 종목코드 : KOSPI, NASDAQ
- fdr.StockListing(종목코드)

```python
df_kospi = fdr.StockListing('KOSPI')
```

```python
df_kospi.shape
```

```python
#  특정 조건 필터링 
# 데이타프레임[조건식]
# 데이타프레임[데이타프레임[컬럼명] 비교연산자 데이타]
df_kospi[df_kospi['Name'] == '삼성전자']
```

```python
df_kospi[df_kospi['Name'] == '삼성전기']
```

```python
# 삼성전기 or 삼성전자 
# 데이타프레임에서의 논리 연산자 
# or => |
# and => & 
df_kospi[(df_kospi['Name'] == '삼성전기') | (df_kospi['Name'] == '삼성전자')]
```

```python
df_kospi['Name']
```

### 특정 문자열이 포함된 데이터
데이터프레임면[컬럼명].str.contains(키워드)

```python
df_kospi['Name'].str.contains('흥국')
```

```python
df_kospi[df_kospi['Name'].str.contains('흥국')]
```

```python
df_kospi.head()
```

```python
# 데이터프레임 행 슬라이싱
df_kospi.loc[1:6]
```

```python
df_kospi.iloc[:3]
```

```python
# 컬럼 슬라이싱
df_kospi.loc[:,'Symbol':'Name'].head()
```

```python
# 행, 컬럼 슬라이싱
df_kospi.loc[1:20, 'Market':'Sector']
```
