# 색상 정보 자료 정리

- https://www.w3schools.com/colors/colors_names.asp


## BeautifulSoup 작업 과정 

- requests 또는 urllib 이용해서 url => html 화 => 텍스트 
- 텍스트 => BeautifulSoup 객체화 
- BeautifulSoup 제공함수를 이용해서 자료 변환 
- csv 파일로 저장


```python
# url => html 문자열화
url = 'https://www.w3schools.com/colors/colors_names.asp'
response = requests.get(url)
html_str= response.text
with open('output/colors_names.html','w') as f:
    f.write(html_str)
```

```python
# BeautifulSoup 객체화
soup = BeautifulSoup(html_str,'html.parser')
soup
```

```python
# 추출하고자 하는 칼라박스 필터링  
# find_all(class_='클래스명') => 리스트 
# .innerbox
div_list = soup.find_all(class_='innerbox')
len(div_list)
```

```python
# 첫번째 요소 - Alice blue
div_list[0]
```

```python
# 마지막 요소 - yellowgreen
div_list[-1]
```

```python
# 색상명, 16진수색상값, url 추출
print(div_list[-1].find('a'))
print(div_list[-1].find(class_='colornamespan').text)  # YellowGreen
print(div_list[-1].find(class_='colorhexspan').text)  # #9ACD32
base_url = 'https://www.w3schools.com'
print(base_url+div_list[-1].find('a')['href'])
```

```python
# 모든 색상 정보를 리스트에 저장하기 => 2차원 구조
color_list = []
base_url = 'https://www.w3schools.com'
for div in div_list:
    color_name = div.find(class_='colornamespan').text
    hexcode = div.find(class_='colorhexspan').text
    url = base_url+div.find('a')['href']
    # 1차원 리스트 구조로 생성한 후 다시 리스트에 저장 => 2차원
    color_list.append([color_name, hexcode, url])
```

```python
# 데이터프레임 형태로 저장
field_list = ['Name', 'Hex', 'URL']
df = pd.DataFrame(color_list, columns=field_list)
df.head()
```

```python
# CSV 파일로 저장 
df.to_csv('output/color_info.csv', index=False)
```
